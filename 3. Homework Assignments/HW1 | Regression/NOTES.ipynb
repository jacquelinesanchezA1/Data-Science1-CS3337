```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame with a target variable 'output'
data = {'output': [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]}
df = pd.DataFrame(data)

# Plot the counts of the target variable
plt.figure(figsize=(8, 6))
df['output'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.xlabel('Output')
plt.ylabel('Count')
plt.title('Counts of Target Variable')
plt.xticks(rotation=0)
plt.show()
```
