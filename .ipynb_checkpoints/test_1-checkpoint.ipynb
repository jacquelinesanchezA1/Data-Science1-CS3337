```python
// Introduction:
// This program demonstrates basic input and output operations in Java.
// It covers printing to the console, reading user input, formatted output,
// and basic file handling operations.

import java.util.Scanner;
impot java.io.FileWriter;
import java.io.FileReader;
import java.io.BufferedReader;
import java.io.IOException;
import java.util.NoSuchElementException;

public class Main {
    public static void main(String[] args) {
        // *Basic Input and Output*

        // Objective: Demonstrate how to print a simple message to the console
        System.out.println("Hello, World!");

        // Objective: Show how to read input from the console and print a personalized message
        Scanner scanner = new Scanner(System.in);
        System.out.print("Enter your name: ");
        
        try {
            if (scanner.hasNextLine()) {
                String name = scanner.nextLine();
                System.out.println("Hello, " + name);
            } else {
                System.out.println("No input provided.");
            }
        } catch (NoSuchElementException e) {
            System.out.println("Error reading input: " + e.getMessage());
        }

        // *Formatted Output*

        // Objective: Display how to use printf for formatted output
        int age = 21;
        System.out.printf("I am %d years old.%n", age);

        // Objective: Demonstrate formatted output using String.format
        int score = 90;
        System.out.println(String.format("Your score is %d", score));

        // *File Input and Output*

        // Objective: Illustrate how to write a string to a file
        try {
            FileWriter writer = new FileWriter("output.txt");
            writer.write("This is a sample, creating and writing on output");
            writer.close();
        } catch (IOException e) {
            e.printStackTrace();
        }

        // Objective: Show how to read the content from a file and print it
        try {
            BufferedReader reader = new BufferedReader(new FileReader("output.txt"));
            String line;
            while ((line = reader.readLine()) != null) {
                System.out.println(line);
            }
            reader.close();
        } catch (IOException e) {
            e.printStackTrace();
        }

        scanner.close();
    }
}
```
